```python
excelname = "kd"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname) + ".xlsx")
ws = wb[str(sheetname)]

dataset = []

i = 1
while(ws.cell(row = i, column = 1).value != None):
    row = []
    j = 1
    while(ws.cell(row = i, column = j).value != None):
        row.append(ws.cell(row = i, column = j).value)
        j = j +1
    dataset.append(row)
    i = i + 1
```

```python
pd.DataFrame(dataset).head(10)
```

```python
# trade =[[0 for x in range(5)] for y in range(len(dataset))]
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 11):
        row.append(0)
    trade.append(row)
pd.DataFrame(trade)  
trade[0][0] = "date"
trade[0][1] = "open"
trade[0][2] = "high"
trade[0][3] = "low"
trade[0][4] = "close"
trade[0][5] = "rsv"
trade[0][6] = "k"
trade[0][7] = "d"
trade[0][8] = "buy_day"
trade[0][9] = "cum_buy_day"
trade[0][10] = "sell_day"
pd.DataFrame(trade).head()
```

```python
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]
    trade[i][1] = dataset[i][1]
    trade[i][2] = dataset[i][2]
    trade[i][3] = dataset[i][3]
    trade[i][4] = dataset[i][4]
pd.DataFrame(trade).head(10)
```

```python
for i in range(9,len(trade)):
    high = trade[i][2]
    low = trade[i][3]
    for j in range(1,9):
        if(trade[i-j][2] > high):
            high = trade[i-j][2]
        if(trade[i-j][3] < low):
            low = trade[i-j][3]
        trade[i][5] = (trade[i][4] - low) / (high - low) * 100
    rsv = trade[i][5] 
pd.DataFrame(trade).head(20)
```

```python
i = 10
trade[9][6] = 50
trade[9][7] = 50
for i in range(10, len(trade)): 
    trade[i][6] = trade[i-1][6] * (2/3) + trade[i][5] * (1/3)   
    trade[i][7] = trade[i-1][7] * (2/3) + trade[i][6] * (1/3)
pd.DataFrame(trade).head(25)
```

```python
for i in range(10, len(trade)):
    if(trade[i][6] > trade[i][7]) and (trade[i-1][8] == 0):
        trade[i][8] = 1
        trade[i][9] = 1
        trade[i][10] = 0
    elif(trade[i][6] < trade[i][7]) and (trade[i-1][8] == 1):
        trade[i][8] = 0
        trade[i][9] = 0
        trade[i][10] = 1
    elif(trade[i-1][8] > 0):
        trade[i][8] = 0
        trade[i][9] = 1
        trade[i][10] = 0   
pd.DataFrame(trade).head(25)
```
